## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
#json_normalize(data) #by me
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### Find the top 10 major project themes and fill in missing themecode

```python
# Load json as strings
data_json=json.load((open('data/world_bank_projects.json')))

# Flattern json
df_json=json_normalize(data_json, 'mjtheme_namecode', ['id', 'countryname'])
df_json.head()
```

```python
# Build a dictionary for the theme code and theme name
code_name_dict=df_json[df_json['name']!=''].set_index('code')['name'].to_dict()

# Fill missing theme names by looking up the code_name_dict
df_json['themename']=df_json.code.replace(code_name_dict)

# Rearrange and rename columns
df_json.drop('name',axis=1,inplace=True)
df_json=df_json[['countryname','id','themename','code']]
df_json.columns=['Country','ProjectId','ThemeName','ThemeCode']
df_json.head()
```

```python
# Check if there is missing value
df_json.isnull().sum()

# Drop duplicated data
#df.duplicated(keep='first')
df_dropduplicate=df_json[['ProjectId','Country']].drop_duplicates(keep='first')

assert (df_dropduplicate.ProjectId.value_counts()==1).sum()==len(df_dropduplicate)
```

```python
# Find the 10 countries with most projects
df_dropduplicate.Country.value_counts()[:10]
```

```python
# Find the top 10 major project themes 
df_json.ThemeName.value_counts()[:10]
```

```python
print('top 10 major project themes :\n\n',df_json.ThemeName.value_counts()[:10].index.values.tolist())

print('number of themes: \n\n',df_json.ThemeName.value_counts()[:10].values)
```
